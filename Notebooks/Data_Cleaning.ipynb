#### Demographic info about customers:

* gender – Customer’s gender (Male/Female).
* SeniorCitizen – Whether the customer is a senior (1) or not (0).
* Partner – Whether the customer has a partner (Yes/No).
* Dependents – Whether the customer has dependents (children or others).

#### Customer account information

* tenure – Number of months the customer has been with the company.
* Contract – Type of contract (Month-to-month, One year, Two year).
* PaperlessBilling – Whether billing is paperless (Yes/No).
* PaymentMethod – Method used for payments (electronic check, mailed check, etc.).
* MonthlyCharges – Monthly amount charged to the customer.
* TotalCharges – Total amount paid by the customer over their entire tenure.

#### Services that each customer has signed up for

* PhoneService – Whether the customer has phone service.
* MultipleLines – Whether they have multiple phone lines.
* InternetService – Type of internet service (DSL, Fiber optic, None).
* OnlineSecurity – Whether online security add-on is enabled.
* OnlineBackup – Whether cloud backup service is enabled.
* DeviceProtection – Whether device protection plan is subscribed.
* TechSupport – Whether technical support add-on is enabled.
* StreamingTV – Whether streaming TV service is active.
* StreamingMovies – Whether streaming movies service is active.
